```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = "listingsutf.csv"
airbnb_df = pd.read_csv(file, error_bad_lines=False)
airbnb_df.head()
```

```python
list(airbnb_df.columns.values)
```

```python
airbnb_df_new = airbnb_df[["neighbourhood","neighborhood_overview", "latitude", "longitude", "property_type","bathrooms", "room_type",
                           "accommodates", "bedrooms", "beds", "price", "number_of_reviews", "review_scores_rating"]].copy()

airbnb_df_new.head()
```

```python
airbnb_df_sqrft = airbnb_df[["neighbourhood","neighborhood_overview", "latitude", "longitude", "property_type","bathrooms", "room_type",
                           "accommodates", "bedrooms", "beds","square_feet", "price", "number_of_reviews", "review_scores_rating"]].copy()
airbnb_df_sqrft.head()
```

```python
airbnb_df_sqrft = airbnb_df_sqrft.dropna()
airbnb_df_sqrft.head()
```

```python
airbnb_df_new = airbnb_df_new.rename(columns={"review_scores_rating": "rating"})
airbnb_df_new.head()
```

```python
airbnb_df_new["price"]= airbnb_df_new["price"].str.replace('$', '')
airbnb_df_new.head()
```

```python
airbnb_df_sqrft["price"]= airbnb_df_sqrft["price"].str.replace('$', '')
airbnb_df_sqrft.head()
```

```python
#rds_connection_string = "root:shayan1371@127.0.0.1/airbnb_db"
#engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
engine = create_engine("sqlite:///airbnb.sqlite")
```

```python
engine.table_names()
```

```python
airbnb_df_new.to_sql(name='airbnb', con=engine, if_exists='append', index = False)
airbnb_df_sqrft.to_sql(name='airbnb_sqrft', con=engine, if_exists='append', index = False)
```

```python
pd.read_sql_query("SELECT latitude FROM airbnb", con = engine).head()
```
